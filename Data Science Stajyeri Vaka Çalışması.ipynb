# 
Kan Grubu eksik verilerini doldururken, anne ve babanın kan grubu bilgilerini göz önünde bulundurmak çok önemli. Kan grupları genetik olarak aktarıldığı için, anne ve babanın kan grupları, çocuklarının kan grubunu sınırlı bir aralıkta tahmin etmemize yardımcı olabilir. Kan grubu mirası, ABO ve Rh sistemlerine bağlıdır.

Kan Grubunun Genetik Geçişi:
A ve B genleri baskındır (dominant), O geni ise çekinik (recessive) bir gendir.
Rh+ geni baskındır, Rh- geni çekiniktir.
Örneğin:

A ve B genleri olan bir kişi AB kan grubuna sahip olur.
A ve O genleri olan bir kişi A kan grubuna sahip olur (çünkü A geni baskındır).
O ve O genleri olan bir kişi O kan grubuna sahip olur (çünkü O çekiniktir ve başka baskın bir gen yoktur).
Kan Grubu Kombinasyonları:
Eğer Anne A ve Baba A ise, çocuk A veya O kan grubuna sahip olabilir.
Eğer Anne A ve Baba B ise, çocuk A, B, AB, veya O kan grubuna sahip olabilir.
Eğer Anne O ve Baba O ise, çocuk O kan grubuna sahip olur.
Rh faktörü de benzer şekilde miras alınır. Hem anne hem baba Rh+ ise, çocuk büyük olasılıkla Rh+ olacaktır (ama Rh- olma ihtimali de vardır).

```python
# Kan grubu kombinasyonlarına göre olası çocuk kan gruplarını tahmin eden fonksiyon
def tahmin_kan_grubu(anne_kan_grubu, baba_kan_grubu):
    # Eğer anne veya baba kan grubu eksikse None döndür
    if pd.isnull(anne_kan_grubu) or pd.isnull(baba_kan_grubu):
        return None

    # ABO sistemi için olası kombinasyonlar
    possible_blood_types = []
    if 'A' in anne_kan_grubu and 'A' in baba_kan_grubu:
        possible_blood_types = ['A', 'O']
    elif ('A' in anne_kan_grubu and 'B' in baba_kan_grubu) or ('B' in anne_kan_grubu and 'A' in baba_kan_grubu):
        possible_blood_types = ['A', 'B', 'AB', 'O']
    elif 'A' in anne_kan_grubu and 'O' in baba_kan_grubu or 'O' in anne_kan_grubu and 'A' in baba_kan_grubu:
        possible_blood_types = ['A', 'O']
    elif 'B' in anne_kan_grubu and 'B' in baba_kan_grubu:
        possible_blood_types = ['B', 'O']
    elif 'B' in anne_kan_grubu and 'O' in baba_kan_grubu or 'O' in anne_kan_grubu and 'B' in baba_kan_grubu:
        possible_blood_types = ['B', 'O']
    elif 'AB' in anne_kan_grubu and 'A' in baba_kan_grubu or 'A' in anne_kan_grubu and 'AB' in baba_kan_grubu:
        possible_blood_types = ['A', 'B', 'AB']
    elif 'AB' in anne_kan_grubu and 'B' in baba_kan_grubu or 'B' in anne_kan_grubu and 'AB' in baba_kan_grubu:
        possible_blood_types = ['A', 'B', 'AB']
    elif 'AB' in anne_kan_grubu and 'AB' in baba_kan_grubu:
        possible_blood_types = ['A', 'B', 'AB']
    elif 'AB' in anne_kan_grubu and 'O' in baba_kan_grubu or 'O' in anne_kan_grubu and 'AB' in baba_kan_grubu:
        possible_blood_types = ['A', 'B']
    elif 'O' in anne_kan_grubu and 'O' in baba_kan_grubu:
        possible_blood_types = ['O']

    # RhD sistemi için olasılıklar
    rh_anne = '+' if 'RH+' in anne_kan_grubu else '-'
    rh_baba = '+' if 'RH+' in baba_kan_grubu else '-'

    if rh_anne == '-' and rh_baba == '-':
        rh_factor = 'RH-'
    else:
        rh_factor = np.random.choice(['RH+', 'RH-'], p=[0.75, 0.25])  # %75 RH+, %25 RH- olasılığı

    # ABO ve RhD gruplarını birleştir
    if possible_blood_types:
        return np.random.choice(possible_blood_types) + " " + rh_factor
    else:
        return None

# Kan grubu doldurma fonksiyonu
def kan_grubu_doldur(row):
    if pd.isnull(row['Kan Grubu']):
        olasi_kan_grubu = tahmin_kan_grubu(row['Anne Kronik Hastaliklari'], row['Baba Kronik Hastaliklari'])
        if olasi_kan_grubu:
            return olasi_kan_grubu
    return row['Kan Grubu']
```

### Kilo & Boy

```python
from sklearn.impute import SimpleImputer

# Eksik verileri ortalama ile doldurma (SimpleImputer ile)
imputer = SimpleImputer(strategy='mean')

# Kilo ve Boy sütunlarında eksik verileri dolduruyoruz
df['Kilo'] = imputer.fit_transform(df[['Kilo']])
df['Boy'] = imputer.fit_transform(df[['Boy']])
```

Veri setinde sayısal sütunlarda eksik veriler bulunmaktaydı. Bu eksik verileri doldurmak için ortalama ile doldurma (mean imputation) yöntemini tercih ettik. Bu karar, verinin dağılımı ve merkezi eğilim ölçütlerine dayanarak alındı.

Özellikle Kilo ve Boy sütunlarını incelediğimizde, eksik verilerin ortalama ile doldurulmasının mantıklı olduğunu gördük. Çünkü:

Kilo sütunundaki verilerin dağılımına baktığımızda: Ortalaması: 80.86 kg Medyanı (50. persentil): 83 kg %25'lik dilim: 65 kg %75'lik dilim: 96 kg Bu değerler, ortalamanın veri dağılımını iyi bir şekilde temsil ettiğini ve %50'lik dilimin ortalama değere oldukça yakın olduğunu gösteriyor. Dolayısıyla, eksik verileri ortalama ile doldurmak, verinin genel dağılımını bozmayacak bir strateji olarak değerlendirildi. Aynı durum Boy sütunu için de geçerli: Ortalaması: 174.64 cm Medyanı (50. persentil): 176 cm %25'lik dilim: 160 cm %75'lik dilim: 187 cm Boy sütunundaki verilerin de ortalama etrafında dengeli bir dağılım gösterdiği ve medyanın ortalamaya çok yakın olduğu görülüyor. Bu nedenle, eksik verileri ortalama ile doldurmak, veri setinin bütünlüğünü korumak açısından uygun bir yöntem olarak tercih edilmiştir. Ortalama ile doldurma, özellikle verinin normal ya da simetrik bir dağılım gösterdiği durumlarda etkili bir yöntemdir. Bu yüzden, eksik veriler için daha karmaşık bir stratejiye ihtiyaç duyulmadı ve bu basit ama etkili yöntemi uyguladık.
